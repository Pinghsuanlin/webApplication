# **3. Implement Flask API functions**

```python
#Return a list of Crunchbase companies based in New York
#!pip install flask
from flask import Flask, render_template #this will give the same template
app = Flask("cb_ny")
```

```python
@app.route('/')
def show_cb_ny():
    global df
    line = df[df['city'] == 'New York']['name'].tolist()
    return render_template('index.html', result=line)     
    #return output
```

```python
app.run(host='localhost', port=4021)

if __name__ == '__main__':
    app.run(debug=True)
```
